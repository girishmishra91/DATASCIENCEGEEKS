```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline
```

### Generating sin and cos graph using matplotlib

```python
x=np.linspace(0,np.pi,5)
sin_y=np.sin(x)
print(x)
print([i for i in sin_y])
```

```python
x=np.linspace(0,4*np.pi,19)
sin_y=np.sin(x)
plt.plot(x,sin_y)
plt.title("Sine wave")
plt.show()
```

```python
x=np.linspace(0,4*np.pi,19)
cos_y=np.cos(x)
plt.plot(x,cos_y)
plt.title("Cosine wave")
plt.show()
```

```python
x=np.linspace(0,4*np.pi,100)
#defining wave
sin_y=np.sin(x)
cos_y=np.cos(x)
# potting wave
plt.plot(x,sin_y,label="sine")
plt.plot(x,cos_y,label="cosine")
plt.legend(loc="upper right")
plt.show()
```

```python
x=np.linspace(0,4*np.pi,100)
#defining wave
sin_y=np.sin(x)
cos_y=np.cos(x)
# potting wave
plt.plot(x,sin_y,"*")
plt.plot(x,cos_y,"--")
plt.legend(loc="upper right")
plt.show()
```

# Different Types Of Plot

### Line plot

```python
x=np.linspace(0,4*np.pi,100)
y=np.sin(x)
plt.style.use("seaborn-v0_8-bright")
plt.plot(x,y)
plt.title("line plot")
plt.show()
```

### multiple line plots

```python
x=np.linspace(0,4*np.pi,100)
plt.style.use('seaborn-v0_8-deep')
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.figure(figsize=(15,4))
plt.plot(x,y_sin,'*',color="blue",linewidth=2,label="Sin")
plt.plot(x,y_cos,'--',color='red',linewidth=2,label="cos")
plt.legend(loc='upper right')
plt.title("Sine&Cosine Plot")
plt.show()
```

### adjusting the plot axes

```python
x=np.linspace(0,4*np.pi,100)
plt.xlim(0,10)
plt.ylim(-2,2)
sin_y=np.sin(x)
plt.plot(x,sin_y)
plt.title("Sine wave")
plt.show()
```

## avoid adjust graph parameter plt.axis('tight') or plt.axis('equal')

### using .set to combine all of theses commands

```python
x=np.linspace(0,4*np.pi,100)
fig=plt.axes()
fig.plot(x,np.sin(x))
fig.set(xlim=(0,10),ylim=(-2,2),xlabel="x",ylabel="sin(x)",title="sin wave")
fig.plot()
plt.show()
```

# Histogram And Binnings

```python
data=np.random.randn(100)
plt.hist(data)
plt.xlabel("values")
plt.ylabel("frequency")
plt.title("random Histogram")
plt.show()
```

```python
data=np.random.randn(100)
plt.hist(data,bins=30,edgecolor='white',color="blue",alpha=.8)
plt.xlabel("values")
plt.ylabel("frequency")
plt.title("random Histogram")
plt.show()
```

### multiple histogram

```python

x1=np.random.randn(100)
x2=np.random.randn(100)+3
x3=np.random.randn(100)+6
plt.hist(x1,bins=20,alpha=.8)
plt.hist(x2,bins=20,alpha=.8)
plt.hist(x3,bins=20,alpha=.8)
plt.xlabel("values")
plt.ylabel("frequency")
plt.title("Multiple histogram")
plt.show()
```

### 2D Histogram

```python
data=np.random.randn(1000)
data_1=np.random.randn(1000)
plt.hist2d(data,data_1,bins=20,cmap="viridis")
plt.title("2D Histogram")
plt.ylabel("frequency")
plt.xlabel("values")
plt.show()
```

```python
data=np.random.randn(1000)
data_1=np.random.randn(1000)
plt.hist2d(data,data_1,bins=20,cmap="Blues")
plt.title("2D Histogram")
plt.ylabel("frequency")
plt.xlabel("values")
plt.colorbar()
plt.show()
```

# Pie Charts

```python
cars=["Fortuner","Defender","Legender","Harrior","Thar"]
data=[50,40,30,70,100]
fig=plt.figure(figsize=(5,7))
plt.pie(data,labels=cars)
plt.show()
```

```python
cars=["Fortuner","Defender","Legender","Harrior","Thar"]
data=[50,40,30,70,100]
explode=[0,0,0,.1,0]
fig=plt.figure(figsize=(5,7))
plt.pie(data,labels=cars,explode=explode)
plt.show()
```

### Adding Shadows

```python
cars=["Fortuner","Defender","Legender","Harrior","Thar"]
data=[50,40,30,70,100]
explode=[0,0,0,.1,0]
fig=plt.figure(figsize=(5,7))
plt.pie(data,labels=cars,explode=explode,shadow=True)
plt.show()
```

### adding percentage

```python
cars=["Fortuner","Defender","Legender","Harrior","Thar"]
data=[50,40,30,70,100]
explode=[0,0,0,.1,0]
fig=plt.figure(figsize=(5,7))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%1.1f%%')
plt.show()
```

### Adding widget

```python
fig=plt.figure(figsize=(5,7))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%1.1f%%',wedgeprops={'edgecolor':'black','linewidth':2,'antialiased':True})
plt.show()
```

```python
fig=plt.figure(figsize=(5,7))
plt.pie(data,labels=cars,explode=explode,shadow=True,autopct='%1.1f%%',wedgeprops={'edgecolor':'black','linewidth':2,'antialiased':False})
plt.show()
```

# Multiple Subplots

```python
x=np.linspace(0,10,50)
fig,ax=plt.subplots(2,2)
y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin wave')

y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title("cos wave")

y=np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title("tan wave")

y=np.random.randn(100)
ax[1,1].hist(y,bins=40)
ax[1,1].set_title("Random Distribution")
plt.tight_layout()
plt.show()
```

## with loops

```python
plt.subplots_adjust(hspace=.4,wspace=.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize=15,ha='center')
```
